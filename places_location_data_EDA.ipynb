### 송파구

```python
songpa = DATA[(DATA['시도']=='서울특별시') & (DATA['시군구']=='송파구') & (DATA['체육시설업종']=='체력단련장업')]
songpa = songpa[songpa.index!=6966]  # index 6966은 송파구에 있지 않음
len(songpa)
```

```python
# json파일을 깃헙에서 Raw로 볼때 어떤 명칭으로 되어있는지 확인후 색칠하고 싶은 것들만 따로 색칠하기가 가능
# 송파구만 지도에 하이라이트하기 위함
temp = {'시군구':['송파구'],
        '색':[1]}
temp = pd.DataFrame(temp)
```

```python
# 코로플레스를 위한 시/군/구별 좌표 json파일 가져오기
import json
import folium

# 중심 지정
lat = songpa['위도'].mean()
long = songpa['경도'].mean()
m = folium.Map([lat, long], zoom_start=13)

# 송파구만 지도에 하이라이트
geo_json='https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2012/json/municipalities-geo-simple.json'
folium.Choropleth(
    geo_data=geo_json,
    data = temp,
    columns = ['시군구','색'],
    fill_color = 'YlGn', # 'BuGn', 'BuPu', 'GnBu', 'OrRd', 'PuBu', 'PuBuGn', 'PuRd', 'RdPu', 'YlGn', 'YlGnBu', 'YlOrBr', 'YlOrRd'
    key_on = 'feature.properties.name'
).add_to(m)

# 체력단련장업들을 지도위에 표시
for i in songpa.index:
    sub_lat = songpa.loc[i, '위도']
    sub_long = songpa.loc[i, '경도']
    title = i  # 이상치를 index로 알아보기 위해
    #지도에 데이터 찍어서 보여주기
    folium.Marker([sub_lat, sub_long], tooltip = title).add_to(m)

# 지도 저장
m.save(dir+'songpa_gym_mark.html')
m
```

### 제주시
