```python
tour_20_df = tour_20[meaningful_colname_list]

# 결측이 없는 칼럼

null_sum = tour_20_df.isnull().sum()
null_sum.loc[null_sum == 0]

notnull = null_sum.loc[null_sum == 0].index
notnull_df = tour_20[notnull].astype(int)

# 응답자 특성 칼럼 리스트
ind_chr_list = notnull_df.columns[0:10]
ind_chr_df = notnull_df[ind_chr_list]

# 여행 총량 칼럼 리스트
exp_col_list = notnull_df.columns[10:16]
exp_col_df = notnull_df[exp_col_list]

# 인구 특성 칼럼 리스트
dq_col_list = notnull_df.columns[16:]
dq_col_df = notnull_df[dq_col_list]
```

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots(5, 2, figsize = (40, 80))
for i, col in enumerate(ind_chr_list):
    sns.countplot(col, data = ind_chr_df, ax = ax[i // 2, i % 2], order = ind_chr_df[col].value_counts().index)

    for p in ax[i // 2, i % 2].patches:
        height = p.get_height()
        ax[i // 2, i % 2].text(p.get_x() + p.get_width() / 2., height + 30, height, ha = 'center', size = 15)

plt.show()
```

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots(3, 2, figsize = (40, 80))
for i, col in enumerate(exp_col_list):
    sns.countplot(col, data = exp_col_df, ax = ax[i // 2, i % 2], order = exp_col_df[col].value_counts().index)

    for p in ax[i // 2, i % 2].patches:
        height = p.get_height()
        ax[i // 2, i % 2].text(p.get_x() + p.get_width() / 2., height + 30, height, ha = 'center', size = 40)

plt.show()
```

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots(4, 2, figsize = (40, 80))
for i, col in enumerate(dq_col_list):
    sns.countplot(col, data = dq_col_df, ax = ax[i // 2, i % 2], order = dq_col_df[col].value_counts().index)

    for p in ax[i // 2, i % 2].patches:
        height = p.get_height()
        ax[i // 2, i % 2].text(p.get_x() + p.get_width() / 2., height + 30, height, ha = 'center', size = 40)

plt.show()
```
